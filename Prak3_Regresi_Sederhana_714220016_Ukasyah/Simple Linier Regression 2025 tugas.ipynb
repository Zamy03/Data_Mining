```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
v_714220016 = pd.read_csv('regresi_ukasyah.csv')
X = v_714220016.iloc[:, :-1].values
y = v_714220016.iloc[:, 1].values
```

```python
v_714220016.info()
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train_714220016, X_test_714220016, y_train_714220016, y_test_714220016 = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

```python
# Fitting Simple Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
LinRe = LinearRegression()
LinRe.fit(X_train_714220016, y_train_714220016)
```

```python
print("Koefisien (β₁):", LinRe.coef_)
print("Intersep (β₀):", LinRe.intercept_)
```

```python

# Predicting the Test set results
y_714220016 = LinRe.predict(X_test_714220016)
```

```python
# Visualising the Training set results
plt.scatter(X_train_714220016, y_train_714220016, color = 'red')
plt.plot(X_train_714220016, LinRe.predict(X_train_714220016), color = 'blue')
plt.title('Keuntungan vs Biaya Iklan (Training set)')
plt.xlabel('Biaya Iklan')
plt.ylabel('Keuntungan')
plt.show()
```

```python
# Visualising the Test set results
plt.scatter(X_test_714220016, y_test_714220016, color = 'red')
plt.plot(X_train_714220016, LinRe.predict(X_train_714220016), color = 'blue')
plt.title('Keuntungan vs Biaya Iklan (Test set)')
plt.xlabel('Biaya Iklan')
plt.ylabel('Keuntungan')
plt.show()
```
