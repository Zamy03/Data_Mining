```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://id.wikipedia.org/wiki/Olimpiade_Catur'
```

```python
table_class = 'wikitable'
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup)
```

```python
conggres = soup.find_all('table', {'class': table_class})
```

```python
df = pd.read_html(str(conggres))[0]
```
